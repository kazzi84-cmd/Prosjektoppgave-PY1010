```python
# Oppgave a


import pandas as pd                           # Modulen pandas importeres og navngis pd

data = pd.read_excel('support_uke_24.xlsx')   # Filen 'support_uke_24' leses inn i programmet.

u_dag = data['Ukedag'].to_numpy()             # Verdiene i kolonne A i filen legges i en array, kolonnen henvises til ved
                                              # overskriften på kolonnen, 'Ukedag'. Dette gjøres for de tre andre 
kl_slett = data['Klokkeslett'].to_numpy()     # kolonnene på linjene under.

varighet = data['Varighet'].to_numpy()

score = data['Tilfredshet'].to_numpy()
```

```python
# Oppgave b

import matplotlib.pyplot as plt              # Modulen matplotlib.pyplot importeres og navngis plt.

u_dag_list = list(u_dag)                     # Arrayen 'u_dag' fra celle 1 gjøres om til en liste.

mandager = u_dag_list.count('Mandag')        # Listefunksjonen count brukes til å telle antallet av hver ukedag.
tirsdager = u_dag_list.count('Tirsdag')
onsdager = u_dag_list.count('Onsdag')
torsdager = u_dag_list.count('Torsdag')
fredager = u_dag_list.count('Fredag')

navn_u_dag = ['Mandag', 'Tirsdag', 'Onsdag', 'Torsdag', 'Fredag']           # Liste med navn på ukedager.

antall_u_dager = [mandager, tirsdager, onsdager, torsdager, fredager]       # Liste med antallet av hver ukedag hentet
                                                                            # fra tellingen over.


plt.bar(navn_u_dag, antall_u_dager)          # Det dannes et søylediagram, en søyle for hver ukedag,
                                             # høyden på søylene angir antall henvendelser den dagen.

plt.xlabel("Ukedag")                         # x- og y-aksen i diagrammet gis merkelapper.
plt.ylabel("Antall henvendelser")
```

```python
# Oppgave c

liste_varighet = list(varighet)              # Arrayen 'varighet' fra celle 1 gjøres om til en liste.

liste_varighet.sort()                        # Listen sorteres fra minste verdi til største verdi.

korteste = pd.to_timedelta(liste_varighet[0])        # Det første (korteste samtale) og siste (lengste samtale) elementet
lengste = pd.to_timedelta(liste_varighet[-1])        # fra lista hentes ut og gjøres om til en datattype som angir tid.


timer_lengste = lengste.components.hours             # Henter ut antall timer, minutter og sekunder varighet på korteste
minutter_lengste = lengste.components.minutes        # og lengste samtale.
sekunder_lengste = lengste.components.seconds

timer_korteste = korteste.components.hours   
minutter_korteste = korteste.components.minutes 
sekunder_korteste = korteste.components.seconds

print(f'Den minste samtaletiden er {timer_korteste} time(r), {minutter_korteste}\   
 minutt(er) og {sekunder_korteste} sekund(er).')                                      # Resultatene printes til skjerm.

print(f'Den lengste samtaletiden er {timer_lengste} time(r), {minutter_lengste}\
 minutt(er) og {sekunder_lengste} sekund(er).')
```

```python
# Oppgave d

antall_tim = 0                                # Lager variabler til å telle antall minutter, timer, sekunder
antall_min = 0                                # og elementer.
antall_sek = 0
indeks = 0


for elm in varighet:                                 # for-løkke som går igjennom hvert element i arrayen 'varighet' fra    
    tid = pd.to_timedelta(elm)          # celle 1. Hvert element gjøres om til en datatype som angir tid, 
    antall_tim += tid.components.hours               # deretter legges antall timer, minutter og sekunder for hvert 
    antall_min += tid.components.minutes             # element til variablene fra linjene over. Indeksen teller en opp
    antall_sek += tid.components.seconds             # for hver runde i løkka for å gå videre til neste element.
    indeks += 1
    
antall_sek = antall_sek + antall_min*60 + antall_tim*3600      # Den sammenlagte tiden av alle samtalene regnes om til sekunder.
gj_antall_sek = antall_sek / (indeks+1)                        # Gjennomsnittlig antall sekunder ved å dele på totalt antall henvendelser.
gj_antall_min = gj_antall_sek // 60                            # Finner antall hele minutter i gjennomsnittet.
rest_sek = round(gj_antall_sek % 60, 2)                        # Finner resterende antall sekunder i gjennomsnittet når hele minutter trekkes fra.



print(f'Gjennomsnittlig samtaletid er {gj_antall_min} minutt(er) og {rest_sek} sekund(er).')       # Resultatet printes til skjerm.
```

```python
# Oppgave e

åtte_ti = 0                                          # Variabler for å tellle antall henvendelser i hvert tidsrom.
ti_tolv = 0                                        
tolv_to = 0
to_fire = 0



for tp in kl_slett:                                  # for-løkke som går igjennom hvert element i arrayen 'kl_slett' fra celle 1.
                                                     # Hvert element gjøres om til en datatype som angir tid. Deretter hentes 
    tidsp = pd.to_timedelta(tp)                      # time-komponenten ut av tiden og disse blir behandlet i if/elif testene der det 
    time = tidsp.components.hours                    # plusses på henvendelser i de respektive variablene for å telle antall henvendelser
                                                     # i hvert tidsrom.
    if time < 10:                                    
        åtte_ti += 1
    elif time > 10 and time < 12:
        ti_tolv += 1
    elif time > 12 and time < 14:
        tolv_to += 1
    elif time > 14:
        to_fire += 1
    


import numpy as np                                  # Modulen numpy importeres og navngis np
import matplotlib.pyplot as plt                     # Modulen matplotlib.pyplot importeres og navngis plt.

tidsrom = np.array([åtte_ti, ti_tolv, tolv_to, to_fire])                                  # Array med antall hendelser per tidsrom.
tekst = np.array(['08:00 - 10:00', '10:00 - 12:00', '12:00 - 14:00', '14:00 - 16:00'])    # Array med forklarende tekst.

plt.pie(tidsrom, labels = tekst, autopct = "%1.1f%%")        # Lager sektordiagram med antall hendvendelser fordelt på tidsrom.
plt.title('Antall henvendelser fordelt på tidsrom')          # Navngir diagrammet.
```

```python
# Oppgave f

neg = 0                                            # Variabler til å telle antall tilbakemeldinger de gitte kategoriene.
nøy = 0
pos = 0


for tall in score:                                 # for-løkke som sjekker verdien av en og en tilbakemelding og teller
    if tall <= 6:                                  # antallet som hører til i hver kategori.
        neg += 1
    elif tall == 7 or tall == 8:
        nøy += 1
    elif tall >= 9:
        pos += 1
        
total = neg + nøy + pos                           # Finner det totale antallet tilbakemeldinger som er kommet inn.
prosent_neg = (neg / total) * 100                 # Finner prosentandel av negative tilbakemeldinger.
prosent_pos = (pos / total) * 100                 # Finner prosentandel av positive tilbakemeldinger.

nps = round( prosent_pos - prosent_neg , 2)       # Finner nps-verdien.

print(f'Supportavdelingens NPS er {nps}.')        # Resultatet printes til skjerm.
```
